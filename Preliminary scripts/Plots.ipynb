```python
#Participación de valor agregado por actividad económica  (porcentaje). participacion_valor_agregado_por_grupo_plot
participacion_valor_agregado_por_grupo_plot, ax = plt.subplots(figsize=(10, 8))
participacion_valor_agregado_por_grupo = participacion_valor_agregado_por_grupo.sort_values(by='Porcentaje', ascending=True)
participacion_valor_agregado_por_grupo.plot(
    kind='barh', 
    x='Nombre_Grupo', 
    y='Porcentaje', 
    color=['#2596be', '#8fcae2'], 
    width=0.4,  # Grosor de las barras reducido
    ax=ax,
    edgecolor='none'  # Desactivar bordes de las barras
)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.set_title('Distribución de valor agregado por actividad económica', fontsize=14)
ax.set_xlabel('Porcentaje (%)', fontsize=12)
ax.set_ylabel('')
for p in ax.patches:
    x = p.get_width()
    y = p.get_y() + p.get_height() / 2
    if x > 0:
        ax.text(x + 0.5, y, f'{x:,.1f}%', ha='left', va='center', fontsize=10, color='black')
plt.grid(False)
plt.tight_layout()
participacion_valor_agregado_por_grupo_plot.show()
```
